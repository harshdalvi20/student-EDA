### 5. Data Transformation


```python
#Add a new column AverageScore which is the mean of MathScore, ScienceScore, and EnglishScore.
df["Average"]=df[["MathScore","ScienceScore","EnglishScore"]].mean(axis=1)
#axis=1 makes the row-wise calculation and axis=0 makes the column-wise calculation
```

```python
df
```

### Add a new column Grade based on the AverageScore:
## A: AverageScore >= 90
## B: 80 <= AverageScore < 90
## C: 70 <= AverageScore < 80
## D: AverageScore < 70

```python
df["Grade"]=df.loc[(df.Average>=90),["Grade"]]="A"
```

```python
df.loc[(df.Average<70),["Grade"]]="D"
```

```python
df.loc[(df.Average>=70) & (df.Average<80),["Grade"]]="C"
```

```python
df.loc[(df.Average>=80) & (df.Average<90),["Grade"]]="B"
```
